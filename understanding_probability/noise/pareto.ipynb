```python
#def pareto(size, a_KB_is_1024_Bytes = True):
def paretocdf(x, xm = 1, xmax = 1, step = 1, alpha=1.6):    
    print(xm, alpha, "\n")
    #x=np.arange(xm,xmax,step,dtype=float)
    y= 1 - (xm/x)**alpha
    mean = alpha * xm/(alpha - 1)
    median = xm * 2**(1/alpha)
    #print(x,"\n")
    #print(np.size(y))
    #plt.plot(x/max(x) * 100,y)
    plt.show()
    plt.xscale("log")
    plt.plot(x,y)
    plt.xlabel("outcomes")
    plt.ylabel("CDF (fraction of wealth)")
    
    return median, mean
```

```python
def paretopdf(x, xm = 1, xmax = 1, step = 1, alpha=1.6):    
    print(xm, alpha, "\n")
    #x=np.arange(xm,xmax,step,dtype=float)
    y= alpha * xm**alpha/x**(1 + alpha)
    mean = alpha * xm/(alpha - 1)
    median = xm * 2**(1/alpha)
    #print(x,"\n")
    #print(np.size(y))
    #plt.plot(x/max(x) * 100,y)
    plt.show()
   # plt.xscale("log")
    plt.yscale("log")
    plt.plot(x,y)
    plt.xlabel("outcomes")
    plt.ylabel("PDF (wealth)")
    
    return median, mean
```

```python
def gaussiancdf(x, mean=0, var=1):
    y = norm.cdf(x, mean, var)
    plt.xscale("log")
    plt.plot(x,y)
    plt.xlabel("outcomes")
    plt.ylabel("CDF (fraction of wealth)")
```

```python
def gaussianpdf(x, mean=0, var=1):
    y = norm.pdf(x, mean, var)
    #plt.xscale("log")
    plt.yscale("log")
    plt.plot(x,y)
    plt.xlabel("outcomes")
    plt.ylabel("PDF (fraction of wealth)")
```

```python
#pareto(1,20, .1)
xm = 1
xmax = 1e9
step = 1e3
#x=np.arange(xm,xmax,step,dtype=float)
x=np.arange(1e8,xmax,step,dtype=float)
median, mean = paretocdf(x,xm,xmax,step)
gaussiancdf(x, mean, 200 * mean)
plt.legend(['Pareto', 'Gaussian'])
#gaussiancdf(x, mean, .1 * mean)
```

```python
#pareto(1,20, .1)
xm = 1
#xmax = 1e13 * 3
xmax = 1e9
step = 1e6
x=np.arange(1,xmax,step,dtype=float)
median, mean = paretopdf(x,xm,xmax,step)
stddev=1e12 * mean
gaussianpdf(x, mean, stddev)
plt.legend(['Pareto', 'Gaussian'])
print("Mean:", mean, "STDDEV_GAUSS:", stddev, "MEDIAN", median)
#gaussianpdf(x, mean, .1 * mean)
```
