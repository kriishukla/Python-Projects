```python
"""Plot Bernoulli trials.
    
    Args:
        n_trials (int): Number of trials.
        p (float): Probability of success for each trial.
"""

import numpy as np
import matplotlib.pyplot as plt
import random

def plot_bernoulli_trials(n_trials, p):

    trials = np.random.binomial(1, p, size=n_trials)
    x = np.arange(1, n_trials+1)

    y = np.cumsum(trials)

    plt.plot(x, y, linestyle='--', marker='o')

    plt.xlabel('Trial Number')

    plt.ylabel('Number of Successes')

    plt.title('Bernoulli Trials')

    plt.show()
    
plot_bernoulli_trials(n_trials=20, p=0.2)


# display the plot
plt.show()


# simulate 20 coin tosses
outcomes = [random.choice(['Heads', 'Tails']) for _ in range(20)]

# create a step plot of the outcomes
plt.step(range(1, 21), outcomes, where='post')

# add axis labels and a title
plt.xlabel('Trial Number')
plt.ylabel('Outcome')
plt.title('Outcomes of 20 Coin Tosses')

# display the plot
plt.show()


"""The first function replicates Bernoulli trials with a particular probability 
of success and graphs the cumulative number of successes versus the trial number.
 The second function generates and plots the results of random processes.
  The second function replicates coin tosses and plots the results in a step-like manner.

The first function is helpful for analysing the behaviour of random processes 
with binary results, such as coin tossing, dice rolling, or determining if a 
binary event is successful or unsuccessful. The law of large numbers, which 
states that the average of the outcomes approaches the expected value as the
 number of trials grows, can be seen by showing the cumulative number of successes 
 across a large number of trials.

A discrete-time stochastic process, such as a series of coin tosses, stock 
prices, or weather measurements, can be visualised using the second function. 
The step plot displays any patterns or trends in the data and demonstrates the 
distinct character of the operation.

Overall, this code offers an easy-to-use method for creating and visualising random 
processes, and it can be simply modified to explore a wide range of phenomena in
 disciplines like physics, finance, and statistics.


"""





```

```python
"""Plot Bernoulli trials multiple times.
    
    Args:
        n_trials (int): Number of trials.
        p (float): Probability of success for each trial.
        n_repeats (int): Number of times to repeat the trials.
"""
def plot_bernoulli_trials_multiple(n_trials, p, n_repeats):

    for i in range(n_repeats):
        trials = np.random.binomial(1, p, size=n_trials)
        x = np.arange(1, n_trials+1)

        y = np.cumsum(trials)

        plt.plot(x, y, linestyle='-', marker='o', label=f'Repeat {i+1}')

    plt.xlabel('Trial Number')

    plt.ylabel('Number of Successes')

    plt.title(f'Bernoulli Trials (n={n_trials}, p={p})')
    
    plt.legend()


    plt.show()
plot_bernoulli_trials_multiple(n_trials=20, p=0.2, n_repeats=5)


# simulate 5 sets of 20 coin tosses
outcomes_list = [[random.choice(['Heads', 'Tails']) for _ in range(20)] for _ in range(5)]

# create a step plot of the outcomes for each set
for i, outcomes in enumerate(outcomes_list):
    plt.step(range(1, 21), outcomes, label=f'Repeat {i+1}', where='post')

# add axis labels, a title, and a legend
plt.xlabel('Trial Number')
plt.ylabel('Outcome')
plt.title('Outcomes of 20 Coin Tosses, 5 Repeats')
plt.legend()

# display the plot
plt.show()
"""In the first piece of code, a function called plot_bernoulli_trials_multiple 
is defined, which generates multiple sets of Bernoulli trials (coin flips) with 
a specified number of trials and success probability. The total number of successes 
for each trial in each set is then plotted. The function requires three arguments: t
the number of trials in each set (n_trials), the chance that each trial will succeed 
(p), and the number of repetitions (n_repeats) needed to generate and plot all the sets.

Using np.random.binomial, which creates a binomial random variable with n_trials 
trials and probability of success p, the function creates each set of trials. 
The results are then plotted using plt.plot after computing the cumulative sum 
of successes using np.cumsum.


The second code snippet mimics five sets of twenty coin flips by randomly 
choosing between "Heads" and "Tails" for each flip using Python's built-in 
random.choice function. It then uses plt.step to generate a step plot of the 
results for each set and plt.xlabel, plt.ylabel, plt.title, and plt.legend to 
add axis labels, a title, and a legend to the plot. Finally, it uses plt.show to display the plot.
"""
```

```python
"""Plot histogram of total number of successes across multiple sequences of Bernoulli trials.
    
    Args:
        n_trials (int): Number of trials in each sequence.
        p (float): Probability of success for each trial.
        n_sequences (int): Number of sequences to generate."""

def plot_bernoulli_histogram(n_trials, p, n_sequences):

    total_successes = []

    for i in range(n_sequences):

        trials = np.random.binomial(1, p, size=n_trials)

        total_successes.append(np.sum(trials))

    plt.hist(total_successes, bins=n_trials+1, density=False)
    
    plt.xlabel('Total Number of Successes')
    plt.ylabel('Frequency')
    plt.title(f'Bernoulli Trials (n={n_trials}, p={p})')
    plt.ylim(0, n_sequences)  # set y-axis limits
    plt.show()

plot_bernoulli_histogram(n_trials=20, p=0.2, n_sequences=100)

"""Plotting the Bernoulli histogram requires three arguments: n_trials, an integer 
reflecting the number of trials in each series, p, a float representing the chance 
of success for each trial, and n_sequences, an integer indicating the number of sequences to construct.

The function creates n_sequences sequences of n_trials Bernoulli trials with a 
success probability of p and calculates the total number of successes in each sequence.
 Then, it sets the y-axis boundaries to be between 0 and n_sequences, plots a histogram 
 of the total number of successes over all sequences with n_trials+1 bins.

The distribution of the total number of successes across numerous sequences of
 Bernoulli trials with the same probability of success is displayed in the following
  histogram. Due to the law of large numbers, it is predicted that the histogram's form 
  will be centred around the expected value of the total number of successes, which is 
  equal to n_trials * p, and that its spread will narrow as the number of sequences rises.

"""

```

```python
# Simulate the sum of normal RVs for each value of n
for n in num_rvs:
    # Generate n sequences of normal RVs with mean mu and standard deviation sigma
    normal = np.random.normal(mu, sigma, size=(n, n))
    # Sum the RVs across each sequence
    sums = np.sum(normal, axis=1)
    # Compute the mean and standard deviation of the sums
    mean = np.mean(sums)
    std = np.std(sums)
    # Append the results to the corresponding lists
    mean_list.append(mean)
    std_list.append(std)

# Plot the means and standard deviations as a function of the number of RVs being summed
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
ax[0].plot(num_rvs, mean_list, 'bo-')
ax[0].set_xlabel('Number of RVs being summed')
ax[0].set_ylabel('Mean of sum')
ax[0].set_title(f'Sum of Normal RVs ($\mu$={mu}, $\sigma$={sigma})')
ax[1].plot(num_rvs, std_list, 'ro-')
ax[1].set_xlabel('Number of RVs being summed')
ax[1].set_ylabel('Standard deviation of sum')
ax[1].set_title(f'Sum of Normal RVs ($\mu$={mu}, $\sigma$={sigma})')
plt.tight_layout()
plt.show()

# Plot a histogram of the sum of normal RVs for n=1000
n = 1000
normal = np.random.normal(mu, sigma, size=(n, n))
sums = np.sum(normal, axis=1)
fig, ax = plt.subplots()
ax.hist(sums, bins=50)
ax.set_xlabel('Sum of normal RVs')
ax.set_ylabel('Frequency')
ax.set_title(f'Histogram of sum of {n} normal RVs')
plt.show()

"""This code generates two random values of alpha uniformly and randomly from 
the interval [1, 20]. For each value of alpha, the code simulates the sum of Poisson r
andom variables (RVs) over an increasing sequence of the number of RVs being added 
(5, 20, 100, 1000 in this case). The law of large numbers and the central limit theorem
 are demonstrated by calculating and plotting the means and standard deviations of the 
 sums for both values of alpha in a 2x2 grid of subplots.

The first part of the code initializes the necessary libraries, sets the seed for reproducibility, 
generates the two random values of alpha, and defines the sequence of the number of RVs being added.

The second part of the code simulates the sum of Poisson RVs and calculates the means and 
standard deviations of the sums for both values of alpha, then appends the results to the corresponding lists.

The third part of the code plots the means and standard deviations as a function of the number
 of RVs being summed, for both values of alpha. The means are plotted with blue circles ('bo-'),
  while the standard deviations are plotted with red circles ('ro-'). The x-axis represents the
   number of RVs being summed, while the y-axis represents the mean or standard deviation of the sum. 
   The title of each subplot indicates the value of alpha for the corresponding Poisson RVs.

Overall, this code demonstrates how the law of large numbers and the central limit theorem can be applied to the sum of Poisson RVs, and how the means and standard deviations of the sum converge to a normal distribution as the number of RVs being summed increases.

"""
```
